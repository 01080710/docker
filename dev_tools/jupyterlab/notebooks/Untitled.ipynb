```python
import tkinter as tk
from tkinter import ttk, messagebox
import sqlite3
from datetime import datetime

# 建立 GUI 視窗
root = tk.Tk()
root.title("時間範圍資料查詢")
root.geometry("600x400")

# 建立資料庫連線（可改成你的資料庫）
conn = sqlite3.connect("example.db")
cursor = conn.cursor()

# 建立範例資料表（可略過這段）
cursor.execute("""
    CREATE TABLE IF NOT EXISTS logs (
        id INTEGER PRIMARY KEY,
        message TEXT,
        timestamp DATETIME
    )
""")
conn.commit()

# 預設一些測試資料（避免重複）
cursor.execute("SELECT COUNT(*) FROM logs")
if cursor.fetchone()[0] == 0:
    import random
    import datetime as dt
    now = dt.datetime.now()
    for i in range(10):
        ts = now - dt.timedelta(hours=random.randint(1, 100))
        cursor.execute("INSERT INTO logs (message, timestamp) VALUES (?, ?)",
                       (f"Log {i}", ts))
    conn.commit()

# ===== 介面元件 =====
tk.Label(root, text="起始時間 (YYYY-MM-DD HH:MM:SS):").pack(pady=5)
start_entry = tk.Entry(root, width=40)
start_entry.pack()

tk.Label(root, text="終止時間 (YYYY-MM-DD HH:MM:SS):").pack(pady=5)
end_entry = tk.Entry(root, width=40)
end_entry.pack()

result_box = tk.Listbox(root, width=80)
result_box.pack(pady=10, expand=True, fill='both')

def query_data():
    result_box.delete(0, tk.END)
    start_time = start_entry.get()
    end_time = end_entry.get()
    
    # 檢查時間格式
    try:
        st = datetime.strptime(start_time, "%Y-%m-%d %H:%M:%S")
        et = datetime.strptime(end_time, "%Y-%m-%d %H:%M:%S")
    except ValueError:
        messagebox.showerror("錯誤", "請輸入正確的時間格式")
        return

    # 查詢資料
    cursor.execute("SELECT id, message, timestamp FROM logs WHERE timestamp BETWEEN ? AND ?", (start_time, end_time))
    rows = cursor.fetchall()
    
    if not rows:
        result_box.insert(tk.END, "沒有資料符合條件")
    else:
        for row in rows:
            result_box.insert(tk.END, f"[{row[2]}] {row[1]} (ID: {row[0]})")

# 查詢按鈕
tk.Button(root, text="查詢", command=query_data).pack(pady=5)

root.mainloop()
```
